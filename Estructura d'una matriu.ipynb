* Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

Primer mostrarem un exemple de broadcast entre un array d'una dimensió i un array de dos dimensions: 


```python
import numpy as np 
# array de dos dimensions: 
a = np.array([[1,2,3], [4,5,6]])
# array d'una dimensió: 
b = np.array([1,2,3])
# ara sumarem les dues arrays: 
a + b
```

Amb el resultat veiem que el array d'una dimensió s'ha trasmés a través de la  segona fila de l'array de 2 dimensions creant una copia per tenir dos dimensions. Veiem que el resultat és el mateix que si haguèssim fet: 

```python
a = np.array([[1,2,3], [4,5,6]])
b2 = np.array([[1,2,3],[1,2,3]])
a+b
```

Ara farem broadcast entre arrays de dos dimensions. Aquesta operació només la  podrem fer si les dues matrius 
tenen la mateixa forma en totes les dimensions o una d'elles té el tamany 1 en la dimensió que no és compatible.  Farem un exemple de dos matrius de dos dimensions, aquest cop farem servir nombres aleatoris: 

```python
c = np.random.randint(1,11, size = (3,3))
d = np.random.randint(1,11, size= (3,1))
print(c)
print(d)
c+d
```

Veiem que la matriu d , que té tamany 1 en una de les seves dimensions, s'ha trasmés per abarcar totes les dimensions de la matriu c.

Ara veurem un exemple entre dos matrius de dos dimensions que no es pot fer broadcast perquè
tenen formes diferents i cap d'elles té tamany 1 en les dimensions no compatibles. 

```python
e = np.random.randint(1,11, size = (2,3))
f = np.random.randint(1,11, size = (3,2))
print(e)
print(f)
e+f 
```

En aquest cas no es pot fer broadcast, no es compleix cap de les dues regles de broadcast: ni tenen la matexia forma en les dues dimensions ni el tamany de les dimensions no compatibles és 1.

Mostreu-me a veure què passa quan eliminem el canal G Verd o B Blau. Hauries d'utilitzar la indexació per seleccionar el canal que voleu anul·lar.

```python
# importem els mòduls: 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# amb imread() carreguem la imatge, veiem que ens torna una matriu 
img = mpimg.imread('city.jpg')
img
```

```python
# si volem veure la imatge: 
imgplot = plt.imshow(img)
```

Si volem separar els canals de cada color, podem seleccionar cada color mitjançcant indexació: 
* vermell = img[:,:,0]
* verd = img[:,:,1]
* blau = img[:,:,2]

```python
# Per treure el canal verd faríem: 
img = mpimg.imread('city.jpg')
img[:, :, 1] = 0
# guardarem la imatge amb el nom city_without_green amb el metode imsave() i la mostrarem per veure com queda: 
plt.imsave("city_without_green.jpg", img)
imgplot = plt.imshow(img)
```

```python
# per treure el canal blau faríem: 
img = mpimg.imread('city.jpg')
img[:,:,2] = 0
# ara guardarem la imatge amb el nom city_without_blue
plt.imsave("city_without_blue.jpg", img)
imgplot = plt.imshow(img)
```
